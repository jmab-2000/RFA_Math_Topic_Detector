<h3>Re-applied RFA on Removed One Value of the Pair of Correlated Values</h3>

```python
import pandas as pd
import numpy as np
import seaborn as sns

data = pd.read_csv('BOW_dataset_v1.csv', encoding='utf-8')
data = data.drop(columns=['Unnamed: 0'])
data.head()
```

```python
correlation = data.corr().abs()
correlation
```

```python
def high_corr(df):
    dataset = pd.DataFrame(columns = ['pair1', 'pair2', 'corr_value'])
    for i in range(len(df)):
        for k in range(i):
            val = df.iloc[i,k]
            if (val >= 0.7):
                dataset = dataset.append({'pair1': df.index[i], 'pair2': df.columns[k], 'corr_value': df.iloc[i,k]}, ignore_index=True)
#                 print(df.index[i]," and " ,df.columns[k], " corr value: ",  df.iloc[i,k])
    return dataset

print("High Correlation Pairs")
hi_corr_table = high_corr(correlation)
# hi_corr_table.to_csv('high_corr_pair.csv')
hi_corr_table
```

```python
drop_cols = hi_corr_table.pair1.unique()
```

```python
# correlation.to_excel('correlation.xlsx')
```

```python
new_data = data
new_data = new_data.drop(columns=drop_cols)

remove_words = pd.DataFrame(drop_cols, columns = ['remove_words'])

# new_data.to_excel('dataset_v2.xlsx')
new_data.shape
```

```python
X = new_data.iloc[:,:len(new_data.columns)-1]
X
```

```python
y = new_data.iloc[:,len(new_data.columns)-1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state= 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100,random_state = 42)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test, y_pred)
```
