<h1>Top Words per class and High Correlated Pairs question instance</h1>

```python
import pandas as pd

data = pd.read_csv('BOW_dataset_v2.csv', encoding='utf-8')
data = data.drop(columns=['Unnamed: 0'])
data.head()
```

separating data by classification

```python
ss_table = data.loc[data['ML_class'] =='series_and_sequence']
p_table = data.loc[data['ML_class'] =='polynomials']
pe_table = data.loc[data['ML_class'] =='polynomial_equation']
```

```python
ss_desc = ss_table.describe()
ss_desc
```

```python
remove_ss = ss_desc.columns[(ss_desc.loc['max'] == 0).tolist()]
print("Count of Columns with Zero Count on series_and_sequence class")
print(len(remove_ss))
print("List of Removed tokens/columns/feature:")
for i in range(len(remove_ss)):
    print(i+1,".) ", remove_ss[i])
```

<h3>Top 20 Most Occuring Words for series_and_sequence</h3>

```python
ss_sum = pd.DataFrame(ss_table.sum(), columns=['count'])
ss_sum = ss_sum.drop(index = remove_ss)
ss_sum = ss_sum.drop(index = ['ML_class'])

print("Top 20 Most Occuring Words for series_and_sequence")
ss_sum.sort_values(by =['count'], axis=0, ascending=False).head(20)
```

```python
print("No of. Features Remaining Series and Sequences: ", len(ss_sum))
```

```python
orig_data = pd.read_csv('BOW_dataset_v1.csv', encoding='utf-8').drop(columns = ['Unnamed: 0'])
corr_pair = pd.read_csv('high_corr_pair.csv').drop(columns = ['Unnamed: 0'])
corr_pair
```

```python
orig_data
```

```python
# High Corr Pairs
pair1 = corr_pair.pair1.tolist()
pair2 = corr_pair.pair2.tolist()
```

```python
raw_dataset = pd.read_csv('initial_dataset.csv', encoding ='utf-8')
```

<h3>High Correlation Word Pairs and the questions that contain the pair</h3>

```python
# look for index of columns that has value of pair 1 and pair 2 from BOW data set

for i in range(len(pair1)):
    print("Pair ", pair1[i], " and ", pair2[i], " corr value = ", corr_pair['corr_value'][i])
    q_index = orig_data[(orig_data[pair1[i]] > 0) & (orig_data[pair2[i]] > 0)].index.tolist()
    print("No. of instances that produces high correlation: ", len(q_index))
    print("first 5 words relating to high correlation pair:")
    with pd.option_context('display.max_colwidth', None,):
        print(raw_dataset.question.loc[q_index])
    print("-----------------------------------------------------")    
```
