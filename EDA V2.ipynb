<h1>Top Words per class and High Correlated Pairs question instance</h1>

```python
import pandas as pd

data = pd.read_csv('BOW_dataset_v2.csv', encoding='utf-8')
data = data.drop(columns=['Unnamed: 0'])
data.head()
```

```python
data.describe()
```

```python
data_table = pd.DataFrame(data.sum(), columns=['count'])
data_table = data_table.drop(index = ['ML_class'])
print("Top 20 Most Occuring Words")
dd_top20 = data_table.sort_values(by =['count'], axis=0, ascending=False).head(20)
dd_top20
```

```python
from matplotlib import pyplot as plt
height = 10
fig, ax = plt.subplots(figsize =(20, height))
topic = 'the corpus'
title = 'Top 20 Words for ' + topic

name = dd_top20.index.values
price = dd_top20['count'].to_list()
# Horizontal Bar Plot
ax.barh(name, price)
 
# Remove axes splines
for s in ['top', 'bottom', 'left', 'right']:
    ax.spines[s].set_visible(False)
 
# Remove x, y Ticks
ax.xaxis.set_ticks_position('none')
ax.yaxis.set_ticks_position('none')
 
# Add padding between axes and labels
ax.xaxis.set_tick_params(pad = 5)
ax.yaxis.set_tick_params(pad = 10)
 
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.5,
        alpha = 0.2)
 
# Show top values
ax.invert_yaxis()
 
# Add annotation to bars
for i in ax.patches:
    plt.text(i.get_width()+0.2, i.get_y()+0.5,
             str(round((i.get_width()), 2)),
             fontsize = 10, fontweight ='bold',
             color ='grey')
 
# Add Plot Title
ax.set_title(title,
             loc ='left', )
 
# Add Text watermark
fig.text(0.9, 0.15, 'Jeeteshgavande30', fontsize = 12,
         color ='grey', ha ='right', va ='bottom',
         alpha = 0.7)
 
# Show Plot
plt.show()
```

```python
orig_data = pd.read_csv('BOW_dataset_v1.csv', encoding='utf-8').drop(columns = ['Unnamed: 0'])
corr_pair = pd.read_csv('high_corr_pair.csv').drop(columns = ['Unnamed: 0'])
corr_pair
```

```python
orig_data
```

```python
# High Corr Pairs
pair1 = corr_pair.pair1.tolist()
pair2 = corr_pair.pair2.tolist()
```

```python
raw_dataset = pd.read_csv('initial_dataset.csv', encoding ='utf-8')
```

<h3>High Correlation Word Pairs and the questions that contain the pair</h3>

```python
# look for index of columns that has value of pair 1 and pair 2 from BOW data set
hi_pair = pd.DataFrame(columns = ['pair1', 'pair2', 'corr_value', 'count'])
for i in range(len(pair1)):
    print("Pair ", pair1[i], " and ", pair2[i], " corr value = ", corr_pair['corr_value'][i])
    q_index = orig_data[(orig_data[pair1[i]] > 0) & (orig_data[pair2[i]] > 0)].index.tolist()
    print("No. of instances that produces high correlation: ", len(q_index))
    hi_pair = hi_pair.append({'pair1': pair1[i], 'pair2': pair2[i], 'corr_value':corr_pair['corr_value'][i], 'count':len(q_index)}, ignore_index=True)
    print("first 5 words relating to high correlation pair:")
    with pd.option_context('display.max_colwidth', None,):
        print(raw_dataset.question.loc[q_index])
    print("-----------------------------------------------------")    
```

```python
hi_pair.loc[hi_pair['count'] >=10].to_csv('hi_pair.csv', encoding='utf-8')
```
