```python
import pandas as pd

data = pd.read_csv('BOW_dataset_v2.csv', encoding='utf-8')
data = data.drop(columns=['Unnamed: 0'])
data.head()
```

```python
ss_table = data.loc[data['ML_class'] =='series_and_sequence']
p_table = data.loc[data['ML_class'] =='polynomials']
pe_table = data.loc[data['ML_class'] =='polynomial_equation']
```

```python
ss_desc = ss_table.describe()
ss_desc
```

```python
remove_ss = ss_desc.columns[(ss_desc.loc['max'] == 0).tolist()]
print("Count of Columns with Zero Count on series_and_sequence class")
print(len(remove_ss))
print("List of Removed tokens/columns/feature:")
for i in range(len(remove_ss)):
    print(i+1,".) ", remove_ss[i])
```

```python
ss_sum = pd.DataFrame(ss_table.sum(), columns=['count'])
ss_sum = ss_sum.drop(index = remove_ss)
ss_sum = ss_sum.drop(index = ['ML_class'])

print("Top 20 Most Occuring Words for series_and_sequence")
ss_sum.sort_values(by =['count'], axis=0, ascending=False).head(20)
```

```python
print("No of. Features Remaining Series and Sequences: ", len(ss_sum))
```
